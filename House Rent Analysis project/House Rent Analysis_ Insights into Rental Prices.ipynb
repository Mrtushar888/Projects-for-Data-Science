# Get Average,Median,Highest,Lowest Rent

```python
print("The Average Rent is",round(df['Rent'].mean()))
print("The Median Rent is",round(df['Rent'].median()))
print("The Highest Rent is",round(df['Rent'].max()))
print("The Minimum Rent is",round(df['Rent'].min()))
```

# Top 10 highest rent and cities

```python
print("\033[1m                Top 10 Highest Rent:\033[0m")
print(df[['City','Area Locality','BHK','Rent']].sort_values(by='Rent',ascending=False).head(10))
```

# Top 10 Minimum Rent and cities

```python
print("\033[1m                Top 10 Minimum Rent:\033[0m")
print(df[['City','Area Locality','BHK','Rent']].sort_values(by='Rent',ascending=True).head(10))
```

# Scatter Plot on House Rents vs House Sizes and City

```python
plt.figure(figsize=(14, 6))
plt.ticklabel_format(style='plain')
sns.scatterplot(data=df, x='Size', y='Rent', hue='City', s=120, marker='s')
plt.title('Size vs Rent vs City', fontweight='bold')
plt.grid(True)
plt.show()

```

# Bar Plot for Cities house count

```python
No_of_house=df['City'].value_counts()
colors = ['pink', 'blue', 'green', 'orange', 'Black','Red']
func=No_of_house.plot(kind='bar',color=colors)
for p in func.patches:
    func.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.xlabel('City',fontweight='bold')
plt.ylabel('Number of Houses',fontweight='bold')
plt.show()
```

# Tenant Preference According to Furnishing Status

```python
grouped=df.groupby('Tenant Preferred')
population_sum = grouped['Furnishing Status'].value_counts()
print(population_sum)
```

# Bar Plot on Different Types of Tenant Preferred

```python
preffered=df['Tenant Preferred'].value_counts()
mac=preffered.plot(kind='bar',color='pink')
for p in mac.patches:
    mac.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```

# Barplot Tenant Preference VS Rent

```python
ab = sns.barplot(data=df,x='Tenant Preferred', y='Rent',ci=None)
for p in ab.patches:
    ab.annotate(int(p.aget_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```

#  Percentage of furnishing status

```python
per_un=df['Furnishing Status'].value_counts()
per_un
```

```python
percentage_furnished = round((per_un['Furnished'] / per_un.sum()) * 100)
percentage_semi_furnished = round((per_un['Semi-Furnished'] / per_un.sum()) * 100)
percentage_unfurnished = round((per_un['Unfurnished'] / per_un.sum()) * 100)

X = [percentage_furnished, percentage_semi_furnished, percentage_unfurnished]
labels = ['Furnished','Semi-Furnished','Unfurnised']
print("Percentage of Furnished:", percentage_furnished, '%')
print("Percentage of Semi-Furnished:", percentage_semi_furnished, '%')
print("Percentage of Unfurnished:", percentage_unfurnished, '%')
```

# Pie chart for furnishing status 

```python
plt.pie(X,labels=labels,autopct='%.2f%%')
plt.title('Furnishing Status')

plt.show()
```

# BarPlot Furnishing Status

```python
found=df['Furnishing Status'].value_counts().plot(kind='bar',color='grey')
plt.grid(True)
for p in found.patches:
    found.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```

# Distribution of different number of BHK available in the Dataset

```python
plt.figure(figsize=(14,6))
Bhk=df['BHK'].value_counts().plot(kind='bar',color='purple')
for p in Bhk.patches:
    Bhk.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```

# Heatmap on BHK vs Area Type

```python
colormap=plt.cm.plasma
sns.heatmap(pd.crosstab(df["Area Type"],
                        df["BHK"]),
                        cmap = colormap);
```

# Barplot city ,rent and area type

```python
plt.figure(figsize=(20,6))
me=sns.barplot(data=df,x='City',y='Rent',hue='Area Type', ci=None)
for p in me.patches:
    height = p.get_height()
    if not np.isnan(height):
        plt.text(p.get_x() + p.get_width() / 2., height, int(height), ha='center', va='bottom', color='black')

plt.show()
```
